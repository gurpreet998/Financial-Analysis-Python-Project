```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
```

```python
df=pd.read_csv("C:\Python Projects\Financial Analysis SimpliLearn\Finance_data.csv")
df.head()
```

```python
df=df.drop(['Mutual_Funds','Equity_Market','Debentures','Government_Bonds','Fixed_Deposits','PPF','Gold'],axis=1)
```

```python
df.head(3)
```

```python
df=df.replace(r'^\s*$',np.nan,regex=True)
```

```python
df.head()
```

```python
sns.countplot(x='gender', data=df , linewidth=1,hue='gender',edgecolor='Black',legend=False)
plt.show()
```

```python
sns.countplot(x='age',data=df,palette='Set3',linewidth=2,edgecolor='black',hue='age',legend=False)
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(x='gender',data=df,hue="Investment_Avenues",linewidth=2,edgecolor='Black',palette='summer')
plt.title('INVESTMENT AVENUES')

plt.subplot(1,2,2)
sns.countplot(x='gender',data=df,hue="Stock_Marktet",linewidth=2,edgecolor='Black',palette='hot')
plt.title('STOCK MARKET')
plt.show()
```

```python
sns.countplot(x='Factor',data=df,edgecolor='black',palette='coolwarm',linewidth=2,hue='Factor',legend=False)
plt.title('Factor Wise Count')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.pointplot(x='Purpose',y='age',data=df,linestyle='--', capsize=.3,color='green')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sns.countplot(hue='gender',x='Duration',palette='viridis',data=df,linewidth=2,edgecolor='black')
plt.title("DURATION")

plt.subplot(1,2,2)
sns.countplot(hue='gender',x='Invest_Monitor',palette='seismic',data=df,linewidth=2,edgecolor='black')
plt.title("INVESTMENT MONITOR")
plt.show()
```
